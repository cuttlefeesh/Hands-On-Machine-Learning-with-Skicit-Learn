# Chapter 1: The Machine Learning Landscape

## Apa itu Pembelajaran Mesin?

Pembelajaran Mesin (Machine Learning/ML) adalah ilmu dan seni untuk mengembangkan sistem yang dapat belajar dari data, tanpa diprogram secara eksplisit. Konsep dasarnya adalah memberi komputer kemampuan untuk mengidentifikasi pola dalam data dan membuat prediksi atau keputusan berdasarkan pola tersebut, tanpa instruksi manual.

## Mengapa Menggunakan Pembelajaran Mesin?

Pembelajaran Mesin dapat digunakan dalam berbagai situasi yang memiliki masalah kompleks, yang mungkin tidak dapat diselesaikan dengan metode tradisional atau jika solusi yang ada membutuhkan banyak aturan dan penyempurnaan. Pembelajaran mesin juga dapat menyesuaikan diri dengan lingkungan yang berubah dan menghasilkan wawasan dari data yang sangat besar.

## Contoh Aplikasi Pembelajaran Mesin

- **Klasifikasi Gambar**: Seperti mengenali objek dalam gambar atau video menggunakan jaringan saraf konvolusional (CNN).
- **Deteksi Tumor**: Menggunakan segmentasi semantik untuk mendeteksi tumor di gambar medis.
- **Klasifikasi Teks**: Mengklasifikasikan komentar forum atau artikel berita.
- **Pengenalan Suara**: Menggunakan data suara untuk membuat aplikasi berbasis suara seperti asisten virtual.
- **Proyeksi**: Seperti memprediksi harga rumah berdasarkan berbagai faktor ekonomi dan demografi.

## Tipe Sistem Pembelajaran Mesin

Pembelajaran mesin dapat dibagi menjadi beberapa kategori berdasarkan cara pelatihan dan penggunaannya:

1. **Pembelajaran Terawasi (Supervised Learning)**: 
   - Menggunakan data yang sudah dilabeli untuk mengajarkan model cara memprediksi hasil (contoh: klasifikasi dan regresi).
   - Algoritma umum: **Linear Regression**, **K-Nearest Neighbors**, **Decision Trees**, **Support Vector Machines (SVM)**.

2. **Pembelajaran Tidak Terawasi (Unsupervised Learning)**:
   - Tidak menggunakan label dalam data. Tujuannya adalah untuk menemukan struktur atau pola yang tersembunyi.
   - Algoritma umum: **Clustering (K-means, DBSCAN)**, **Dimensionality Reduction (PCA)**.

3. **Pembelajaran Semi-Terawasi (Semisupervised Learning)**:
   - Kombinasi data terlabel dan tidak terlabel, di mana sedikit data terlabel digunakan untuk melatih model.

4. **Pembelajaran Penguatan (Reinforcement Learning)**:
   - Sistem belajar dengan cara mencoba berbagai tindakan, mengamati hasilnya, dan berusaha untuk mengoptimalkan strategi tindakannya untuk mendapatkan penghargaan.

## Pembelajaran Batch dan Online

- **Pembelajaran Batch**: Data pelatihan diproses sekaligus dalam satu batch.
- **Pembelajaran Online**: Sistem dapat belajar secara terus menerus saat data baru datang, memungkinkan adaptasi cepat terhadap data yang berubah.

## Pembelajaran Berdasarkan Instansi vs Model

- **Pembelajaran Berdasarkan Instansi**: Sistem belajar berdasarkan contoh yang ada, dan menggunakan pengukuran kemiripan untuk membuat prediksi.
- **Pembelajaran Berdasarkan Model**: Sistem belajar dengan membangun model dari data yang ada dan menggunakannya untuk prediksi.

## Tantangan Utama dalam Pembelajaran Mesin

1. **Kuantitas Data yang Tidak Cukup**: Data yang terlalu sedikit untuk membuat model yang baik.
2. **Data Tidak Representatif**: Data yang tidak mencakup berbagai situasi atau kasus yang sebenarnya.
3. **Data Berkualitas Buruk**: Data yang banyak noise atau kesalahan yang mengurangi akurasi model.
4. **Fitur yang Tidak Relevan**: Penggunaan fitur yang tidak terkait dengan hasil yang ingin diprediksi.
5. **Overfitting**: Model terlalu spesifik untuk data pelatihan sehingga tidak dapat di-generalize dengan baik.
6. **Underfitting**: Model terlalu sederhana sehingga tidak dapat menangkap pola dalam data.

## Evaluasi dan Validasi Model

Untuk memastikan model akan bekerja dengan baik pada data baru, model harus dievaluasi menggunakan **data uji** yang tidak digunakan selama pelatihan. Evaluasi ini akan memberikan gambaran tentang seberapa baik model menggeneralisasi dan memprediksi data yang belum pernah dilihat.
